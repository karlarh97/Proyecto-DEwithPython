#### Como aplicar una funciona a una columna de pandas

```python
import os
import pandas as pd
```

```python
sales_df = pd.read_csv("sales.csv")
```

```python
sales_df.head()
```

```python
### 20201219 -> 2020/12/19
### 20201219 -> 19-12-2020
sales_df.dtypes
```

```python
def transform_date(text):
    text = str(text)
    d = "{}/{}/{}".format(text[0:4],text[4:6],text[6:])
    return d
    
```

```python
sales_df["OrderDate1"] = [transform_date(i) for i in sales_df['OrderDateKey']]
sales_df.head()
```

```python
sales_df["OrderDate2"] = sales_df['DueDateKey'].map(transform_date)
sales_df.head()
```

```python
sales_df["OrderDate3"] = list(map(transform_date, sales_df['DueDateKey']))
sales_df.head()
```

```python
sales_df['OrderDate4'] = sales_df['OrderDateKey'].apply(transform_date)
sales_df.head()
```

```python
sales_df['OrderDate5'] = sales_df['OrderDateKey'].transform(transform_date)
sales_df.head()
```

```python
sales_df['OrderDate6'] = sales_df['DueDateKey'].aggregate(transform_date)
sales_df.head()
```

```python
sales_df['OrderDate7'] = sales_df['ShipDateKey'].agg(transform_date)
sales_df.head()
```

```python
sales_df[['OrderDate8']] = sales_df[['OrderDateKey']].applymap(str).applymap(lambda s: "{}/{}/{}".format(s[0:4],s[4:6],s[6:] ))
sales_df.head()
```

```python
sales_df['OrderDate9'] = sales_df['OrderDateKey'].apply(str).apply(lambda s: "{}/{}/{}".format(s[0:4],s[4:6],s[6:] ))
sales_df.head()
```
